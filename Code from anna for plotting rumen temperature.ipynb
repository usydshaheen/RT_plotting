```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import datetime
```

```python
root_dir = os.getcwd()
f_name = "CAAS2067.csv"
df = pd.read_csv(f'{root_dir}/{f_name}', parse_dates = ["Date"])
```

```python
#Plot daily average rumen temperature for animal
```

```python
mean_by_day = df.groupby('Date')['RumenT'].mean()
plt.plot(mean_by_day, 'o')
plt.title(f'Cattle {f_name.split(".")[0]} : Daily average rumen temperature')
plt.xlabel('Time (days)')
plt.ylabel("Rumen temperture (Celsius)")
plt.show()
```

```python
# Plot rumen temperature of animal per day
```

```python
def my_plot(gr_df, str_date = '2012-11-22 00:00:00'):
    if gr_df.iloc[0].Date == datetime.strptime(str_date, "%Y-%m-%d %H:%M:%S"):
        import matplotlib.ticker as ticker
        tick_spacing = 10
        
        fig, ax = plt.subplots(1,1)
        ax.plot(gr_df.Time.values, gr_df.RumenT.values, "o")
        ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
        plt.xticks(rotation = 30)
        plt.xlabel('Timestamp (10 mins)')
        plt.ylabel("Rumen temperature (Celsius)")
        plt.title(f'Cattel {f_name.split(".")[0]} : Rumen temperature on {str_date.split(" ")[0]}')
mean_by_day = df.groupby('Date').apply(lambda x: my_plot(x))
plt.show()
```
